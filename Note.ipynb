28 对象方法

````python
# 实例化对象时，init会自动运行.
class Student:
    def __init__(self,name,Student_id): # self把属性的值绑定在实例对象上
        self.name = name #左边是绑定到对象上的属性，右边是普通变量，由传入参数决定
        self.student_id = Student_id
        self.grades = {'语文':0,'数学':0,'英语':0} #如果每个对象拥有同样的初始值，就不需要从参数去获取

    def set_grade(self,course,grade): # 方法可以获取or修改，和对象绑定的属性
        if course in self.grades:
            self.grades[course] = grade

    def print_grades(self):
        print(f'学生:{self.name} (学号:{self.student_id}) 的成绩为:')
        for course in self.grades:
            print(f'{course}:{self.grades[course]}分')


chen = Student('小陈','10086')
chen.set_grade('语文',92)
chen.set_grade('数学',94)
chen.print_grades()
# zeng = Student('小曾','10087')
# chen.name,zeng.grades
# zeng.set_grade('数学',95)
# zeng.grades
````

29 类继承：子类会继承父类的属性和方法       
当A是B，classA(B):      
当新能源车是车，class新能源车(车)：

````python
# 父类
class Mammal:
    def __init__(self,name,sex):
        self.name = name
        self.sex = sex
        self.num_eyes = 2

    def breathe(self):
        print(self.name + '在呼吸')

    def poop(self):
        print(self.name + '在拉屎')
````

````python
# 子类
class Human(Mammal):
    def __init__(self,name,sex):
        super().__init__(name,sex)  # 调用父类的的构造函数
        self.has_tail = False       # 不同子类的特有属性

    def read(self):                 # 不同子类的特有方法
        print(self.name + '在阅读')

class Cat(Mammal):
    def __init__(self,name,sex):
        super().__init__(name,sex)
        self.has_tail = True            # 不同子类的特有属性

    def scratch_sofa(self):             # 不同子类的特有方法
        print(self.name + '在抓沙发')

    def poop(self):
        print(self.name + '在猫砂上拉屎') # 若子类和父类有同名方法，则优先调用子类的方法

cat2 = Cat('Jojo', '男')
cat2.scratch_sofa()
cat2.poop()
cat2.has_tail
````

练习：人力系统

````python
class Employee():
    def __init__(self,name,id):
        self.name = name
        self.id = id

    def print_info(self):
        print(f'员工的姓名是:{self.name},员工的工号是:{self.id}')

class FullEmployee(Employee):
    def __init__(self,name,id,monthly_salary):
        super().__init__(name,id)
        self.monthly_salary = monthly_salary

    def calculate_monthly_pay(self):
        return self.monthly_salary

class ParttimeEmployee(Employee):
    def __init__(self,name,id,daily_salary,work_days):
        super().__init__(name,id)
        self.daily_salary = daily_salary
        self.work_days = work_days
    
    def calculate_monthly_pay(self):
        return self.daily_salary * self.work_days

ff = FullEmployee('小明','10086',10000)
pp = ParttimeEmployee('小王','10087',300,22)
````

````python
ff.print_info()
pp.print_info()
ff.calculate_monthly_pay(),pp.calculate_monthly_pay()
````

- 高阶函数：**把函数作为参数的函数**。格式化函数formatter也可以作为参数，扩展出不同打印格式。    
`calculate_print = (3,calculate_square,formatter)` **这里不用带()和参数**, 如果带()就是在调用函数，传入的就是函数的执行结果，而不是函数本身。       
- 匿名函数，不需要起名字，像个一次性物品，即用即扔，适用一次性调用的场景。当成没有名字，且占用行数更少的函数。

>>冒号前面的变量名，表示传给匿名函数的参数num1，num2；冒号后面直接放上需要返回的结果num1+num2。冒号后面只能是单个表达式..       
>>>匿名函数被定义好后，也可以像普通函数般被直接调用，不同点在于**匿名函数本身也要加括号**。

- json有两种数据结构,对象和数组,可以分别被转换成py字典和列表。一个对象是一个示例，对应dataframe的一行
>对象以大括号{}开头和结尾，然后里面都是键值对，每个键值对之间用逗号进行分隔。不要求所有对象的键都相同
>>数组以中括号[]开头和结尾，然后里面是一个个值，每个值之间用逗号进行分隔.

- json有一些额外规则:
>比如python,字典可以用**整数等不可变数据类型**作为键。但json数据里面只能是**字符串作为键，键不能重复,而且字符串必须被双引号包围，单引号不行。**
>>另外，这个规则不仅针对外层的键，如果json对象或数组里面又嵌套了对象，那规则也同样适用于里面的键。  

json值的类型需要属于以下这6种。
- 第一是字符串，注意用双引号包围。
- 第二是整数或浮点数都行。
- 第三是布尔值，也就是true或者false.因为它本质是JavaScript的对象，而JavaScript的布尔值是以小写开头。
- 第四是数组，json对象的值可以是数组，json数组的值也可以是数组。支持嵌套就可以支持复杂的数据格式。
- 第五是对象，对象的值可以是对象;数组里面的值也可以是对象。
- 第六是空值null，你可以把它看成是py的None，总之都是用来表示值不存在.
